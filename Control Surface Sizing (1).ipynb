```python
import numpy as np
import matplotlib.pyplot as plt
```

### Control Surface Sizings

## Aileron Sizing

```python
Swing = 0.356811561911126
bwing = 1.5494
Cwing = 0.25085177299713585

RatioS = np.linspace(0.03,0.12,100)
RatioB = np.linspace(0.2,0.4,100)
RatioC = np.linspace(0.15,0.3,100)
Saileron = RatioS*Swing
Baileron = RatioB*bwing
Caileron = RatioC*Cwing

MaxS = 0.12*Swing
MaxB = 0.4*bwing
MaxC = 0.3*Cwing
MinS = 0.03*Swing
MinB = 0.2*bwing
MinC = 0.15*Cwing

print(f'The Minimum Recommended Aileron Planform is:\n {MinS} meters^2.\n The Maximum Reccomended Aileron Planform is:\n {MaxS} meters^2.')
print(f'The Minimum Recommended Aileron Span is:\n {MinB} meters.\n The Maximum Reccomended Aileron Span is:\n {MaxB} meters.')
print(f'The Minimum Recommended Aileron Chord is:\n {MinC} meters.\n The Maximum Reccomended Aileron Chord is:\n {MaxC} meters.')


```

```python
plt.title("Aileron Surface Area")
plt.xlabel("Aileron Surface Area Ratio(m^2)")
plt.ylabel("Aileron Surface Area")
plt.plot(RatioS,Saileron)
```

```python
plt.title("Aileron Span")
plt.xlabel("Aileron Span Ratio")
plt.ylabel("Aileron Span(m)")
plt.plot(RatioB,Baileron)
```

```python
plt.title("Aileron Chord")
plt.xlabel("Aileron Chord Ratio")
plt.ylabel("Aileron Chord(m)")
plt.plot(RatioC,Caileron)
```

## Elevator Sizing

```python
Scanard = 0.08920289047778149
Bcanard = 0.762
Ccanard = 0.1254258864985679

RatioS = np.linspace(0.15,0.4,100)
RatioB = np.linspace(0.8,1,100)
RatioC = np.linspace(0.2,0.4,100)
Selevator = RatioS*Scanard
Belevator = RatioB*Bcanard
Celevator = RatioC*Ccanard

MaxS = 0.4*Scanard
MaxB = Bcanard
MaxC = 0.4*Ccanard
MinS = 0.15*Scanard
MinB = 0.8*Bcanard
MinC = 0.2*Ccanard

print(f'The Minimum Recommended Elevator Planform is:\n {MinS} meters^2.\n The Maximum Reccomended Elevator Planform is:\n {MaxS} meters^2.')
print(f'The Minimum Recommended Elevator Span is:\n {MinB} meters.\n The Maximum Reccomended Elevator Span is:\n {MaxB} meters.')
print(f'The Minimum Recommended Elevator Chord is:\n {MinC} meters.\n The Maximum Reccomended Elevator Chord is:\n {MaxC} meters.')
```

```python
plt.title("Elevator Surface Area")
plt.xlabel("Elevator Surface Area Ratio(m^2)")
plt.ylabel("Elevator Surface Area")
plt.plot(RatioS,Selevator)
```

```python
plt.title("Elevator Span")
plt.xlabel("Elevator Span Ratio")
plt.ylabel("Elevator Span(m)")
plt.plot(RatioB,Belevator)
```

```python
plt.title("Elevator Chord")
plt.xlabel("Elevator Chord Ratio")
plt.ylabel("Elevator Chord(m)")
plt.plot(RatioC,Celevator)
```
